```python
from matplotlib import pyplot as plt
from predict import AnomalyCLIPConfig, predict_anomalyclip
from PIL import Image
```

## Color Detection

```python
config = AnomalyCLIPConfig(
    query_image_path='./dataset/metal_nut/test/color/016.png',
    kshot=5,
    output_dir="./results/color",
    gamma=2.5
)

results = predict_anomalyclip(config)

if results:
    img_original = Image.open("./dataset/metal_nut/test/color/016.png")
    img_result   = results['pil_image']

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(img_original)
    axes[1].imshow(img_result)
    plt.show()
```
